```python

x_axis = [x for x in range(0,1000,10)]
print(x_axis)
loss_mean = []
loss_mean_attack = []
loss_median_attack = []
loss_trimmed_mean_attack = []

f = open('loss_mean_0.005.txt', 'r')
loss_mean_string = f.readlines()
for element in loss_mean_string:
  temp = float(element[1:-2].split(", ")[1])
  loss_mean.append(temp)
f.close()

f = open('loss_mean_attack_1_0.005.txt', 'r')
loss_mean_attack_string = f.readlines()
for element in loss_mean_attack_string:
  temp = float(element[1:-2].split(", ")[1])
  loss_mean_attack.append(temp)
f.close()

f = open('loss_median_attack_1_0.005.txt', 'r')
loss_median_attack_string = f.readlines()
for element in loss_median_attack_string:
  temp = float(element[1:-2].split(", ")[1])
  loss_median_attack.append(temp)
f.close()

f = open('loss_trimmed_mean_attack_1_0.005.txt', 'r')
loss_trimmed_mean_attack_string = f.readlines()
for element in loss_trimmed_mean_attack_string:
  temp = float(element[1:-2].split(", ")[1])
  loss_trimmed_mean_attack.append(temp)
f.close()


  
```
